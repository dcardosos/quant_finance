# Extraindo os Portfólios mensais e suas composições 
# (10 Ações que tiveram maior volume no mês anterior)

```python
date_ibx = list(ibx.index)
```

```python
def agrupamento_ações(n_ações, coluna = '6. volume'):
    '''
    Retorna as N maiores ações que fizeram parte da composição do ibx em deternimanda data,
    baseado em um recorte feito pelo nome da coluna que se deseja obter tais ações.
    '''
    lista_portifólios = []

    for date in date_ibx:
        portifólio_ações = []
        i = date_ibx.index(date)
        disponíveis = list(set(ibx.iloc[i,:].values)- set(falhas))
        for item in disponíveis:
            if type(item) != str:
                disponíveis.remove(item)
        for w in range(n_ações):
            try:
                portifólio_ações.append(ações[disponíveis].loc[date,(slice(None), coluna)].nlargest(n=n_ações).index[w][0])
            except:
                continue

        lista_portifólios.append(portifólio_ações)

    return  lista_portifólios
```

```python
portfólio_10_volumes = agrupamento_ações(10,coluna = '6. volume')
```

# Retornos do Portfólio com 10 ações

```python
def retornos_dos_portfólios(lista_portfólios, peso = 0.1):
    '''
    Recebe uma lista com a compoisção de cada portfólio 
    Retorna os retornos de cada portifólio mensal dado as suas composições e seus respectivos pesos
    '''
    retornos_estratégia =[]
    #Não temos a última composição do ibx de 2020-01 , porém é necessário o retorno do último portfólio
    datas = date_ibx.copy()
    datas.append('2020-01')
    for i in range(len(lista_portfólios)):
        retorno_ações = matrix_retornos.loc[datas[i+1], lista_portfólios[i]] * peso
        retorno_portfólio_mensal = ((retorno_ações+1).prod()-1).round(4)
        retornos_estratégia.append(retorno_portfólio_mensal)
    
    return np.array(retornos_estratégia)
```

# Resumo Dados Quant para as estratégias

```python
import metricas as mt
```

```python
def resumo_quant(retornos_mensais,index = False):
    
    '''
    Recebe um array com os retornos mensais de uma estratégia 
    Retorno uma série de Métricas
    '''
    #Retornos
    n_meses = len(retornos_mensais)
    retorno_ap = (retornos_mensais+1).prod() -1
    retorno_aa = (retorno_ap+1)**(12/n_meses) - 1
    retorno_médio = (retorno_ap+1)**(1/n_meses) - 1
    #Vol da estartégia
    volatilidade_am =  retornos_mensais.std()
    volatilidade_aa =  retornos_mensais.std() * (12**0.5)
    # Semideviations
    sd = mt.semideviation(retornos_mensais)
    #VaR
    var_hist = mt.var_historic(retornos_mensais) 
    cvar = mt.cvar_historic(retornos_mensais) 
    var_gaussiano = mt.var_gaussian(retornos_mensais,modified=True)
    cor_fish = mt.var_gaussian(retornos_mensais,modified=False)
    if index:
        sharpe = 0
    else:
        sharpe = mt.sharpe_ratio(retornos_mensais,df_ret_ibov.iloc[:,0].values,12)


    resumo = {'Retorno ao Período': retorno_ap, 
    'Retorno Anualizado': retorno_aa,
    'Retorno Médio': retorno_médio,
    'Volatilidade da estratégia ao mês':volatilidade_am,
    'Volatilidade da estratégia anualizada':volatilidade_aa,
    'Semideviation':sd,
    'VaR Histórico':var_hist,
    'CVaR':cvar,
    'VaR Gaussiano': var_gaussiano,
    'Cornish-Fisher': cor_fish,
    'Information ratio (Ibovespa)': sharpe}

    return resumo
```
